### Distplot

```python
df.columns
```

```python
plt.figure()
fig, ax = plt.subplots(4, 4, figsize=(20, 10))
for var, subplot in zip(df.columns, ax.flatten()):
    sns.distplot(df[var], ax=subplot)

#1) alcohol            - Uniform Distribution
#2) Malic_acid         - Skewed Distribution
#3) ash                - Normal Distribution
#4) alcalinity_of_ash  - Normal Distribution
#5) magnesium          - Normal Distribution
#6) total_phenols      - Uniform Distribution
#7) flavanoids            Uniform Distribution
#8) nonflavanoid_phenols - Binomial Distribution
#9) proanthocyanins      - Normal Distribution
#10) color_intensity     - Skewed Distribution
#11) hue                 - Normal Distribution
#12) od315_of_diluted_wines
#13) proline              - Skewed Distribution
```
